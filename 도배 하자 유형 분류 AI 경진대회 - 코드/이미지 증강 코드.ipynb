```python
import random
import numpy as np
import os
import cv2
import glob
from PIL import Image
import PIL.ImageOps
```

```python
num_augmented_images = 500 # 새로 만들 이미지 갯수를 정하는 것
```

## 파일 경로를 계속 바꾸면서 이미지 증강이 필요한 폴더를 채웠음

```python
file_path = 'C:/Users/pc/OneDrive/바탕 화면/4-1/광진구_프로젝트/train/피스//'
file_names = os.listdir(file_path)
total_origin_image_num = len(file_names)
augment_cnt = 1
```

```python
for i in range(1, num_augmented_images):
    change_picture_index = random.randrange(1, total_origin_image_num)
    print(change_picture_index)
    print(file_names[change_picture_index])
    file_name = file_names[change_picture_index]

    origin_image_path = 'C:/Users/pc/OneDrive/바탕 화면/4-1/광진구_프로젝트/train/피스//'+ file_name
    print(origin_image_path)
    image = Image.open(origin_image_path)
    random_augment = random.randrange(1,3)

    if(random_augment == 1):
        #이미지 좌우 반전
        print("invert")
        inverted_image = image.transpose(Image.FLIP_LEFT_RIGHT)
        inverted_image.save(file_path + 'inverted_' + str(augment_cnt) + '.png')
        
    elif(random_augment == 2):
        #이미지 기울이기
        print("rotate")
        rotated_image = image.rotate(random.randrange(-20, 20))
        rotated_image.save(file_path + 'rotated_' + str(augment_cnt) + '.png')
        

        
    augment_cnt += 1
```
